```python
# This notebook mines the structure of "total" figures in consumption data

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fftpack import fft

data_dir = "../Datasets/consumption.csv"
csp = pd.read_csv(data_dir)
```

```python
csp = csp[csp["area"]=="National"]
```

```python
csp = csp[csp["consumption_segment"]=="All"]
```

```python
csp = csp[csp["indicator"]=="Annual household expenditure by product/service"]
```

```python
ppp = csp["ppp_conversion_factor"]
```

```python
data_dir = "../Datasets/health_indicators.csv"
health = pd.read_csv(data_dir)
```

```python
health = health[health["year"]==2010];
```

```python
health[["country_code","Population, total"]];
```

```python
#csp = csp.head(200)
```

```python
csp.groupby('country_code')
csp['buy']=0
for index, row in csp.iterrows():
    cc = row["country_code"]
    prod = row["product"]
    pp = health[health["country_code"]==cc]["Population, total"]
    print(pp)
    a = row['value']/(row['ppp_conversion_factor']*pp.values[0])
    csp.loc[(csp["country_code"]==cc) & (csp["product"]==prod), 'buy'] = a
    print(a)
```

```python
csp[csp["country_code"]=="CHN"]
#csp
```

```python
total = csp.groupby("country_code")["buy"].sum()
```

```python
total.to_csv("../Data/Consumption/total.csv")
```
