```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_dir = "../Datasets/consumption.csv"
csp = pd.read_csv(data_dir)#.head(100)
csp = csp[csp["indicator"]=="Share of product/service in total expenditure (%)"]
csp
```

```python
dic = {}
for index, row in csp.iterrows():
    area = row["area"]
    consumption_segment = row["consumption_segment"]
    product = row["product"]
    country_code = row["country_code"]
    value = row["value"]
    i_index = str(area)+","+str(consumption_segment)+","+str(product)
    if not country_code in dic.keys():
        dic[country_code] = {}
    dic[country_code][i_index] = value
```

```python
con = pd.DataFrame.from_dict(dic)
np.savetxt("../Data/Consumption/consumption_all_feature.txt",con)
with open("../Data/Consumption/consumption_country.txt", 'w') as f:
    for item in con.columns.values:
        f.write("%s\n" % item)
with open("../Data/Consumption/consumption_feature.txt", 'w') as f:
    for item in con.index.values:
        f.write("%s\n" % item)
```
