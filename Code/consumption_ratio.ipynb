```python
############This notebook mines the structure of ratio data in consumption

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fftpack import fft

data_dir = "../Datasets/consumption.csv"
csp = pd.read_csv(data_dir)
```

```python
csp = csp[csp["area"]=="National"]
```

```python
csp = csp[csp["consumption_segment"]=="All"]
```

```python
csp = csp[csp["indicator"]=="Share of product/service in total expenditure (%)"][["country_code","product","value"]]
```

```python
prods = csp["product"].unique()
cc = csp["country_code"].unique()
```

```python
df_ = pd.DataFrame(index=cc, columns=prods)
for index, row in csp.iterrows():
    print(row["value"])
    df_.set_value(row["country_code"], row["product"], row["value"])
```

```python
data = pd.read_csv("../Data/Consumption/ratio.csv").values[:,1:]
```

```python
data = data.astype(float)
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components = 3)
pca.fit(data)
Y = pca.transform(data)
for i in range(3):
    plt.plot(pca.components_[i])
    plt.savefig("../Data/Consumption/Fig/"+"PCA_"+str(i)+".jpg",bbox_inches="tight")
    plt.show()
```

```python
plt.figure(figsize=(20,20))

plt.subplot(1,3,1)
data_rec = pca.inverse_transform(Y)
plt.imshow(data_rec)

plt.subplot(1,3,2)
plt.imshow(data)

plt.subplot(1,3,3)
plt.imshow(data)
```

```python
def normalize(vec):
    mean = np.mean(vec)
    std = np.std(vec)
    return (vec-mean)/std
```
