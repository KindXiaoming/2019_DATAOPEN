```python
# we use nonlinear methods(autoencoder) to extract features out of consumption data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_dir = "../Data/Complete_Merged_Feature/feature_impute.txt"
data = np.loadtxt(data_dir)
csp = data[:,27:]

from sklearn.decomposition import PCA
losses = []

for i in range(40):
    pca = PCA(n_components = i)
    pca.fit(csp)
    Y = pca.transform(csp)
    csp_rec = pca.inverse_transform(Y)
    losses.append(np.sum((csp-csp_rec)**2))
```

```python
plt.scatter(np.arange(40),losses)
```

```python
data_dir = "../Data/Complete_Merged_Feature/feature_merge.csv"
cols = pd.read_csv(data_dir).columns[28:]
rows = pd.read_csv(data_dir)["country_code"]
data_dir = "../Data/Consumption/consumption_country.txt"
with open(data_dir) as f:
    developing = f.read().splitlines()
deing = csp[[rows[i] in developing for i in range(len(rows))]]
```

```python
plt.scatter(deing[:,0],deing[:,6])
```
